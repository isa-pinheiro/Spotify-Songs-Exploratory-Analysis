### Top 5 Most Popular Artist

```python
top5_artists = df.groupby('artist').count().sort_values('song_title', ascending=False)['song_title'][:5]
top5_artists
```

```python
top5_artists.plot.barh()
plt.show()
```

### Top 5 Loudest Tracks

```python
top5_loudest = df.sort_values('loudness', ascending=True)[['song_title', 'loudness']][:5]
top5_loudest
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x='loudness', y='song_title', data=top5_loudest)
plt.title('Top 5 Loudest Tracks')
plt.show()
```

### Artist with the most danceability song

```python
top5_daceability_artist = df[['danceability', 'song_title', 'artist']].sort_values(by='danceability', ascending=False)[:5]
top5_daceability_artist
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x='danceability', y='artist', data=top5_daceability_artist)
plt.title('Artist with the most danceability song')
plt.show()
```

### Top 10 Instrumentalness Tracks

```python
top10_instrumentalness = df[['instrumentalness', 'song_title', 'artist']].sort_values(by='instrumentalness', ascending=False)[:10]
top10_instrumentalness
```

```python
plt.figure(figsize=(12, 7))
plt.pie(x='instrumentalness', data=top10_instrumentalness, autopct='%1.2f%%', labels=top10_instrumentalness.song_title)
plt.title('Top 10 Instrumentalness Tracks')
plt.show()
```

### Multiple Features Plot

```python
interested_cols = ['tempo', 'loudness', 'acousticness', 'danceability', 'duration_ms', 
                  'energy', 'instrumentalness', 'liveness', 'speechiness', 'valence']
```

```python
for feature_col in interested_cols:
    pos_data = df[df['target'] == 1][feature_col]
    neg_data = df[df['target'] == 0][feature_col]

    plt.figure(figsize=(12, 7))

    sns.histplot(pos_data, bins=30, alpha=0.7, label='Positive', color='green', kde=True)
    sns.histplot(neg_data, bins=30, alpha=0.7, label='Negative', color='red', kde=True)
    
    plt.legend(loc='upper right')
    plt.title(f'Positive and Negative Histogram Plots For {feature_col}')
```
